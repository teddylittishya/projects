```python
import requests
from bs4 import BeautifulSoup
import csv
from pprint import pprint

headers =  {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
page = requests.get("https://www.flipkart.com/mobiles-accessories/mobiles/pr?sid=tyy,4io&q=mobile&otracker=categorytree",headers = headers)
print(page)
fd = page.content
pprint(fd)
```

```python
soup = BeautifulSoup(fd,'html.parser')
pprint(soup)
```

```python
content = soup.find_all('div',class_='_2kHMtA')
print(content)
```

```python
itemlist =[]
for item in content:
    item_name = item.find('div',class_='_4rR01T')
    item_price = item.find('div',class_='_30jeq3 _1_WHN1')
    item_rate = item.find('div',class_='_3LWZlK')
    if item_rate is not None:
        item_dict = {'name':item_name.text,'price':item_price.text,'rate':item_rate.text}
        itemlist.append(item_dict)
        print(item_dict)
        #print(item_name.text,item_price.text,item_rate.text)
    #print(itemlist)
  
```

```python
file = open('flipk1.csv','w',encoding='utf-8',newline='')
fieldname=['name','price','rate']

writer=csv.DictWriter(file,fieldnames=fieldname)
writer.writeheader()
writer.writerows(itemlist)
file.flush()
file.close
```
